```python
#count vowels in string
def count_vowels(string):
  for i in string:
    vowels = "aeiouAEIOU"
    count = 0
    for char in string:
        if char in vowels:
            count += 1
    return count
a="samiksha"
print(count_vowels(a))




```

```python
#fibonacci series
n = 10
a = 0
b = 1
next = b
count = 1

while count <= n:
    print(next, end=" ")
    count += 1
    a, b = b, next
    next = a + b
print()
```
